## Thresholding

````python
image = cv2.imread(fname)
````

````python
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
````

````python
jimshow_channel(grey_image)
````

````python
plt.hist(grey_image.flatten(), 256, [0,256])
plt.show()
````

````python
blurred = cv2.blur(grey_image, (5,5))
````

````python
jimshow_channel(blurred)
````

````python
plt.hist(blurred.flatten(), 256, [0,256])
plt.show()
````

```cv2.threshold(image, threshold_value, colour, method)```

````python
(T_value, threshold) = cv2.threshold(blurred, 200, 255, cv2.THRESH_BINARY)
````

````python
jimshow_channel(threshold)
````

````python
(T_value, thres_inverted) = cv2.threshold(blurred, 200, 255, cv2.THRESH_BINARY_INV)
````

````python
jimshow_channel(thres_inverted)
````

````python
jimshow(cv2.bitwise_and(image, image, mask = thres_inverted))
````

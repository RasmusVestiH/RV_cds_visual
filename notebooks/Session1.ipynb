## Image translation

```python
path_to_image = os.path.join("..","data", "img", "trex.png")
image = cv2.imread(path_to_image)
jimshow(image, "original")
```

__Translate__

Shifting an image down and right

```python
#Make a translation matrix
M = np.float64([[1,0,25],
              [0,1,50]])
```

This shifts 25 pixels on the x-axis and 50 on y-axis

```python
width = image.shape[1]
height = image.shape[0]
```

```python
shifted = cv2.warpAffine(image, M, (width, height))
```

```python
jimshow(shifted, "shifted")
```

It is now shifted 25 to the right and 50 pixels down. 

__Shifting back (left and up)__

Shift 100 pixels left and 100 pixels up

```python
#Make a new translation matrix
M = np.float64([[1,0,-100],
              [0,1,-100]])
```

```python
shifted = cv2.warpAffine(image, M, (width, height))
```

```python
jimshow(shifted, "shifted up and left")
```

```python
#What does this do? 
M = np.float64([[1,0,0],
              [0,1,100]])
shifted = cv2.warpAffine(image, M, (width, height))
```

```python
jimshow(shifted, "shifted down 100 pixels")
```

__Writing a function for images__

```python
def translate(image, x, y):
    #Define translation matrix here 
    M = np.float64([[1,0,x],
                  [0,1,y]])
    #Perform translation on our chosen image
    shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
    
    #Return the translated image
    return shifted
```

```python
shifted_image = translate(image, 0, 100) #You can add the parameters:
```

```python
shifted_image = translate(image=image, x=0, y=100) 
```

```python
jimshow(shifted_image, "Shifted down 100 pixels")
```
